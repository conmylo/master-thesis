```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# Create a small synthetic dataset
sequences = [
    "the quick brown fox jumps over the lazy dog",  # User 1
    "the fast brown fox leaps over the lazy dog",   # User 1
    "the quick brown dog jumps over the lazy fox",  # User 2
    "the fast brown dog leaps over the lazy fox",   # User 2
]

# Labels for the users
labels = ["User 1", "User 1", "User 2", "User 2"]

# Encode the sequences to numerical values
tokenizer = tf.keras.preprocessing.text.Tokenizer(char_level=True)
tokenizer.fit_on_texts(sequences)
encoded_sequences = tokenizer.texts_to_sequences(sequences)

# Pad sequences to ensure they are of the same length
max_sequence_len = max([len(seq) for seq in encoded_sequences])
padded_sequences = tf.keras.preprocessing.sequence.pad_sequences(encoded_sequences, maxlen=max_sequence_len)

# Encode labels
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)
```

```python
# Create the LSTM model
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=8, input_length=max_sequence_len))
model.add(LSTM(32))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, encoded_labels, test_size=0.5, random_state=42)

# Train the model
model.fit(X_train, y_train, epochs=10, verbose=1)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {accuracy:.2f}")

# Test with a new sequence (you can replace this with actual user input in practice)
test_sequence = "the quick brown fox jumps over the lazy dog"
encoded_test_sequence = tokenizer.texts_to_sequences([test_sequence])
padded_test_sequence = tf.keras.preprocessing.sequence.pad_sequences(encoded_test_sequence, maxlen=max_sequence_len)

# Predict the user
prediction = model.predict(padded_test_sequence)
user = label_encoder.inverse_transform([int(prediction[0] > 0.5)])[0]
print(f"Predicted User: {user}")
```
