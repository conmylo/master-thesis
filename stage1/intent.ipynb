```python
import rasa

# Load pre-trained Rasa model
interpreter = rasa.nlu.model.Interpreter.load("path/to/pretrained_model")

def classify_intent(text):
    # Predict intent using Rasa model
    result = interpreter.parse(text)
    intent = result.get("intent", {}).get("name")
    return intent

# Example usage
text = "Can you help me with something?"
intent = classify_intent(text)
print("Intent:", intent)
```
