```python
from pynput import keyboard
import time
from statistics import mean

# Define variables to store keystroke dynamics data
keypress_times = {}
key_release_times = {}
flight_times = []
latencies = []
digraph_latencies = []
error_count = 0
typed_prompt = []
start_time = None

# Function to handle key press
def on_press(key):
    global start_time

    try:
        if key == keyboard.Key.enter:
            return False  # Stop listener after pressing Enter

        if start_time is None:
            start_time = time.time()  # Start timer on first key press

        key_name = str(key.char) if hasattr(key, 'char') else str(key)
        keypress_times[key_name] = time.time()

        if len(typed_prompt) > 0:
            prev_key = typed_prompt[-1]
            latency = keypress_times[key_name] - key_release_times.get(prev_key, keypress_times[key_name])
            latencies.append(latency)

            # Record digraph (pair of two characters) latencies
            if len(typed_prompt) > 1:
                digraph_latencies.append(latency)

        typed_prompt.append(key_name)

    except AttributeError:
        pass  # Special keys like shift/ctrl are ignored for now

# Function to handle key release
def on_release(key):
    global error_count

    key_name = str(key.char) if hasattr(key, 'char') else str(key)
    key_release_times[key_name] = time.time()

    if key_name == 'Key.backspace':
        error_count += 1

    # Calculate flight time (time between release and next press)
    if len(typed_prompt) > 1:
        prev_key = typed_prompt[-2]
        flight_time = keypress_times[key_name] - key_release_times.get(prev_key, keypress_times[key_name])
        flight_times.append(flight_time)

# Function to calculate keystroke dynamics features
def calculate_keystroke_features():
    # Dwell time (time between key press and key release)
    dwell_times = [key_release_times[key] - keypress_times[key] for key in keypress_times]
    dwell_time_avg = mean(dwell_times) if dwell_times else 0

    # Flight time (time between releasing one key and pressing the next)
    flight_time_avg = mean(flight_times) if flight_times else 0

    # Average latency between key presses
    latency_avg = mean(latencies) if latencies else 0

    # Digraph latency (time between specific key pairs)
    digraph_latency_avg = mean(digraph_latencies) if digraph_latencies else 0

    # Typing speed (characters per second)
    total_time = time.time() - start_time
    typing_speed = len(typed_prompt) / total_time if total_time > 0 else 0

    # Error rate (number of backspaces)
    global error_count

    return {
        'prompt': ''.join(typed_prompt),
        'dwell_time_avg': dwell_time_avg,
        'flight_time_avg': flight_time_avg,
        'latency_avg': latency_avg,
        'digraph_latency_avg': digraph_latency_avg,
        'typing_speed': typing_speed,
        'error_rate': error_count
    }

# Main function to capture keystrokes
def main():
    print("Start typing your prompt. Press Enter when finished:")

    with keyboard.Listener(on_press=on_press, on_release=on_release) as listener:
        listener.join()

    # Calculate and display the features
    features = calculate_keystroke_features()
    print("\nKeystroke Dynamics Features:")
    for feature, value in features.items():
        print(f"{feature}: {value}")

# Execute the main function
if __name__ == "__main__":
    main()
```

```python
from pynput import keyboard
import time
from statistics import mean

# Define variables to store keystroke dynamics data
keypress_times = {}
key_release_times = {}
flight_times = []
latencies = []
digraph_latencies = []
error_count = 0
typed_prompt = []
start_time = None

# Function to handle key press
def on_press(key):
    global start_time

    try:
        if key == keyboard.Key.enter:
            return False  # Stop listener after pressing Enter

        if start_time is None:
            start_time = time.time()  # Start timer on first key press

        # Handle normal characters
        if hasattr(key, 'char'):
            key_name = key.char
            keypress_times[key_name] = time.time()
            typed_prompt.append(key_name)  # Add typed character to prompt
        # Handle space bar separately
        elif key == keyboard.Key.space:
            key_name = ' '
            keypress_times[key_name] = time.time()
            typed_prompt.append(key_name)  # Add space character to prompt

        # Calculate latency between key presses
        if len(typed_prompt) > 1:
            prev_key = typed_prompt[-2]
            latency = keypress_times[key_name] - key_release_times.get(prev_key, keypress_times[key_name])
            latencies.append(latency)

            # Record digraph (pair of two characters) latencies
            if len(typed_prompt) > 1:
                digraph_latencies.append(latency)

    except AttributeError:
        pass  # Special keys like shift/ctrl are ignored for now

# Function to handle key release
def on_release(key):
    global error_count

    if hasattr(key, 'char'):
        key_name = key.char
        key_release_times[key_name] = time.time()
    elif key == keyboard.Key.space:
        key_name = ' '
        key_release_times[key_name] = time.time()

    if key == keyboard.Key.backspace:
        error_count += 1

    # Calculate flight time (time between release and next press)
    if len(typed_prompt) > 1:
        prev_key = typed_prompt[-2]
        flight_time = keypress_times[key_name] - key_release_times.get(prev_key, keypress_times[key_name])
        flight_times.append(flight_time)

# Function to calculate keystroke dynamics features
def calculate_keystroke_features():
    # Dwell time (time between key press and key release)
    dwell_times = [key_release_times[key] - keypress_times[key] for key in keypress_times]
    dwell_time_avg = mean(dwell_times) if dwell_times else 0

    # Flight time (time between releasing one key and pressing the next)
    flight_time_avg = mean(flight_times) if flight_times else 0

    # Average latency between key presses
    latency_avg = mean(latencies) if latencies else 0

    # Digraph latency (time between specific key pairs)
    digraph_latency_avg = mean(digraph_latencies) if digraph_latencies else 0

    # Typing speed (characters per second)
    total_time = time.time() - start_time
    typing_speed = len(typed_prompt) / total_time if total_time > 0 else 0

    # Error rate (number of backspaces)
    global error_count

    return {
        'prompt': ''.join(typed_prompt),  # Return the typed prompt
        'dwell_time_avg': dwell_time_avg,
        'flight_time_avg': flight_time_avg,
        'latency_avg': latency_avg,
        'digraph_latency_avg': digraph_latency_avg,
        'typing_speed': typing_speed,
        'error_rate': error_count
    }

# Main function to capture keystrokes
def main():
    print("Start typing your prompt. Press Enter when finished:")

    with keyboard.Listener(on_press=on_press, on_release=on_release) as listener:
        listener.join()

    # Calculate and display the features
    features = calculate_keystroke_features()
    print("\nKeystroke Dynamics Features:")
    for feature, value in features.items():
        print(f"{feature}: {value}")

# Execute the main function
if __name__ == "__main__":
    main()
```
