```python
# Step 1: Import necessary libraries
import numpy as np
from tensorflow.keras.datasets import reuters
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Embedding, Conv1D, MaxPooling1D, GlobalMaxPooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Step 2: Load the Reuters dataset
max_words = 10000
maxlen = 100
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=max_words)

# Step 3: Preprocess the data
tokenizer = Tokenizer(num_words=max_words)
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

# Convert class vectors to binary class matrices (one-hot encoding)
num_classes = np.max(y_train) + 1
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)

# Step 4: Load pre-trained GloVe embeddings
embedding_dim = 100
embedding_index = {}
with open('glove.6B.100d.txt', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = coefs

embedding_matrix = np.zeros((max_words, embedding_dim))
for word, i in tokenizer.word_index.items():
    if i < max_words:
        embedding_vector = embedding_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector

# Step 5: Build the enhanced CNN model
model = Sequential()
model.add(Embedding(max_words, embedding_dim, input_length=maxlen, weights=[embedding_matrix], trainable=True))
model.add(Conv1D(128, 5, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(128, 5, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(128, 5, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(128, 5, activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dropout(0.5))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Set up early stopping and model checkpoint
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model_cnn.keras', save_best_only=True, monitor='val_loss')

# Step 6: Train the model
history = model.fit(x_train, y_train,
                    batch_size=128,
                    epochs=20,
                    verbose=1,
                    validation_split=0.2,
                    callbacks=[early_stopping, model_checkpoint])

# Step 7: Evaluate the model
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# List of Reuters topics
reuters_topics = [
    'cocoa', 'grain', 'crude', 'trade', 'interest', 'money-fx', 'money-supply', 
    'ship', 'sugar', 'coffee', 'gold', 'tin', 'strategic-metal', 'livestock', 
    'retail', 'nat-gas', 'copra-coconut', 'linseed', 'fuel-oil', 'veg-oil', 
    'soy-oil', 'soybean', 'iron-steel', 'rubber', 'heat-wood', 'platinum', 
    'cotton', 'rice', 'wheat', 'oilseed', 'meal-feed', 'reserves', 'instal-debt', 
    'carcass', 'fish', 'orange', 'f-cattle', 'lean-hogs', 'rapeseed', 'soy-meal', 
    'silver', 'lumber', 'copper', 'housing', 'income', 'jobs', 'gas', 'sorghum', 
    'palm-oil', 'pet-chem', 'supply-demand', 'zinc', 'bop', 'money-market', 
    'lead', 'potato', 'wool', 'heat', 'hog', 'sugarbeet', 'commodity', 'grain',
    'total', 'cotton', 'palm-oil', 'rumour', 'money-supply', 'money-supply', 
    'tax', 'meat', 'finance', 'instal-debt', 'tin', 'sugar', 'lead', 'indian-oil', 
    'dfl-bond', 'bfr-bond', 'crude', 'food', 'livestock', 'petroleum', 'dfl-bond', 
    'bop', 'rapeseed', 'money-fx', 'fuel-oil', 'veg-oil', 'soy-oil', 'ship', 
    'fishmeal', 'crude', 'dfl-bond', 'jpy-bond', 'nat-gas', 'gold', 'crude'
]

# Step 8: Classify new user input
def classify_text(text):
    # Preprocess the input text
    sequences = tokenizer.texts_to_sequences([text])
    matrix = pad_sequences(sequences, maxlen=maxlen)
    
    # Predict the topic
    prediction = model.predict(matrix)
    topic_idx = np.argmax(prediction)
    confidence = np.max(prediction)
    
    # Get the topic name
    topic_name = reuters_topics[topic_idx] if topic_idx < len(reuters_topics) else 'unknown'
    
    return topic_name, confidence

# Step 9: Get user input and classify
user_input = input("Enter a news article: ")
topic, confidence = classify_text(user_input)
print(f'The predicted topic is "{topic}" with confidence {confidence:.2f}')
```
